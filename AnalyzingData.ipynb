```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\Franco\Desktop\proyecto analisis de datos\movies\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
df['votes'] = df['votes'].fillna(0)
df['score'] = df['score'].fillna(0)
df['runtime'] = df['runtime'].fillna(0)
```

```python
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
NewDataFrame = df["released"].str.split(" ",n=3,expand = True) 
NewDataFrame.rename(columns={0: "Month", 1: "day",2:"Year",3:"Country"}, inplace=True)
df['yearcorrect'] = NewDataFrame['Year']
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross earnings')

plt.xlabel('Budget for film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'orange'}, line_kws={'color': 'green'})
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```
